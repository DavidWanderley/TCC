```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime

pd.set_option("display.max_columns", 50)
pd.set_option("display.max_rows", 50)
```

```python
years = list(range(2009,2020))
years
```

```python
df = {}
for y in years:
    url = 'data/reclamacoes%s.csv' % y
    df[y] = pd.read_csv(url, sep=';', dtype='unicode')
```

```python
df[2016].tail(10)
```

```python
fig = plt.figure(figsize=[20,20])
for i, y in zip(range(1,12), years):
    fig.add_subplot(6, 2, i)
    sns.heatmap(df[y].isnull(), cbar = False);
```

```python
procon_df = pd.concat([pd.DataFrame(df[y]) for y in years],
          ignore_index=True)
```

```python
procon_df.drop(['AnoCalendario', 'CodigoRegiao', 'Regiao', 'strRazaoSocial', 'RadicalCNPJ', 'strNomeFantasia', 'RazaoSocialRFB','NomeFantasiaRFB','DescCNAEPrincipal','DescricaoAssunto','DescricaoProblema','CEPConsumidor'], axis = 1, inplace=True)
```

```python
def calc_dias_diff(dt_final, dt_inicial):
    return (datetime.strptime(dt_final, '%Y-%m-%d %H:%M:%S.%f').date() - datetime.strptime(dt_inicial, '%Y-%m-%d %H:%M:%S.%f').date()).days
```

```python
procon_df['DiasProcessamento'] = procon_df.apply(
    lambda row: calc_dias_diff(row['DataArquivamento'], row['DataAbertura']),
    axis=1
)
```

```python
procon_df.sample(5)
```

```python
procon_df.drop(['DataArquivamento', 'DataAbertura'], axis = 1, inplace=True)
```
